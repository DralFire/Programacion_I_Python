```python
import pyodbc
import pandas as pd
import tkinter as tk

from tkinter import *
from tkinter import ttk #Importamos todas las funciones que contiene tkinter
from tkinter.ttk import *
from tkinter import messagebox

server = 'localhost'
bd = 'db_mysql'
usuario = 'root'
password = 'abc123'

class General:
	def __init__(self, raiz):
		self.genero = StringVar()
		self.label_genero = Label(raiz, text = "Gender")
		self.label_genero.grid(column=0, row=0)
		self.genero = Combobox(raiz, values=('W', 'M'), width=5)
		self.genero.grid(column=0, row=1)

		self.estado = StringVar()
		self.label_estado = Label(raiz, text = "Country")
		self.label_estado.grid(column=0, row=5)
		
        #Creamos el combobox para seleccionar el estado, CAMBIO: Se delimita a 3 países ***
		self.estado = Combobox(raiz, values=(           
            "CHL",            
            "COL",            
            "MEX"
			)
        , width=10)
		self.estado.grid(column=0,row=6)

		#Creamos los botones
		#Con command le decimos cual función queremos que lleve a cabo
		self.boton_buscar= Button(raiz, text="Search", command=self.buscar)
		self.boton_buscar.grid(column=0, row=20)

		self.boton_borrar=Button(raiz, text="Clean", command=self.borrar)
		self.boton_borrar.grid(column=0, row=30)

		#Tabla
		self.tabla=ttk.Treeview(raiz, column=("c1", "c2", "c3"), show='headings', height=8)
		self.tabla.column("# 1",anchor=CENTER, stretch=NO, width=100)
		self.tabla.heading("# 1", text="Name")
		self.tabla.column("# 2", anchor=CENTER, stretch=NO)
		self.tabla.heading("# 2", text="Gender")
		#self.tabla.column("# 3", anchor=CENTER, stretch=NO)
		#self.tabla.heading("# 3", text="Estado Civil")
		self.tabla.grid(column=0, row=40)

	def buscar(self):        
        
		genero_valor = "'" + self.genero.get() + "'"
		print(genero_valor)

		estado = "'" + self.estado.get() + "'"
		print(estado)

		try:
				conexion = pyodbc.connect(driver='{MySQL ODBC 9.4 Unicode Driver}', SERVER = server, DATABASE = bd,UID=usuario,PWD=password, PORT=3306)
				print('Conexión exitosa')
		except:
				print('La conexión no fué exitosa')

		#Creamos un cursor para almacenar la información en memoria
		cursor = conexion.cursor()

		cursor.execute("SELECT Year, Race, Name, Gender, Age, State, Country, Overall, Finish FROM maratonny_python3 WHERE Gender= " + genero_valor + "AND Country = " + estado)

		datos_clientes = cursor.fetchall()
		print(datos_clientes)
		conexion.commit()

		#Nos aseguramos de cerrar la conexión
		conexion.close()
        
		messagebox.showinfo("Results", datos_clientes)

	#def desplegar_resultados(self):

	def borrar(self):
		self.genero.set("")
		self.estado.set("")





#Creamos el objeto que será la raiz de la aplicación
raiz = Tk()
#Le agregamos un título | CAMBIO: Se modifica el título ***
raiz.title("Search My Score in GO Marathon")
#Determinamos si se podrá cambiar su tamaño
raiz.resizable(1,1)
#Asignamos un logotipo | CAMBIO: Se modifica el ícono ***
raiz.iconbitmap('estacion-oro-gasolinera-tijuana.ico')
#Asignamos un tipo de cursor, un color de background y un borde a la raiz
raiz.config(bd=8)
raiz.config(relief="ridge")
estructura = General(raiz)

raiz.mainloop()
```
